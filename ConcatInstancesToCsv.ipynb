```python
#Imports
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import MinMaxScaler
```

```python
#Reused functions to read the profits and weights from the instances 
def read_profit(line):
    line = str(line)
    line = line[(line.find(",")+1):]
    line = line[:line.find(",")]
    return int(line)

def read_weight(line):
    line = str(line)
    line = line[(line.find(",")+1):]
    line = line[(line.find(",")+1):]
    line = line[:line.find(",")]
    return int(line)

#Function that initialized the DataFrame (Column definition here is not necessary but can be used to validate the data)
def initialize_dataframe_columns():
    empty_df = pd.DataFrame(columns=["capacity", "max_profit"]) #First two columns
    for i in range(1,1001):
        
        #Define two columns to add per item
        newcol = pd.DataFrame(columns=["profit_item_"+str(i), "weight_item_"+str(i)])
        
        #Add the two new columns to the DataFrame
        empty_df = pd.concat([empty_df, newcol])
    return empty_df

#Function to create a DataFrame of the two columns "profit" and "weight" out of the instances
def build_items(items):
    for index in items.index: #For each row
        
        #Read the values
        profit = read_profit(items[0][index])
        weight = read_weight(items[0][index])
        
        #Write the values to new columns
        items.at[index, "profit"] = profit
        items.at[index, "weight"] = weight
    return pd.concat([items["profit"], items["weight"]], axis=1) #Remove the read_line column

#Function to scale the items per column: scales profits and weights individually 
def scale_items(items):
    scaled = min_max_scaler.fit_transform(items)
    return pd.DataFrame(scaled, columns=["profit", "weight"])

#Function to write the values into the DataFrame
def write_profit_weight_to_csv(instances_overview, filepath, index, profit, weight):
    instances_overview.at[filepath, "profit_item_"+str(index + 1)] = profit
    instances_overview.at[filepath, "weight_item_"+str(index + 1)] = weight
```

```python
#IMPORTANT NOTE: For this scaling approach the max_profit and capacity are scaled across all instances 
#Define the scaler
min_max_scaler = MinMaxScaler()

#Create the DataFrame
instances_overview = initialize_dataframe_columns()

for filepath in os.listdir("knapsacksolver/dataset/"):
    if "knapPI" in filepath: #Ignore the Jupyter checkpoints
        with open("knapsacksolver/dataset/"+str(filepath)) as file:
            lines = file.readlines()
            
        #Read capacity
        capacity = lines[2]
        capacity = int(capacity[2:])
        
        #Read max_profit
        max_profit = lines[3]
        max_profit = int(max_profit[2:])
        
        #Write capacity and max_profit to the DataFrame (these values are unscaled)
        instances_overview.at[filepath, "capacity"] = capacity
        instances_overview.at[filepath, "max_profit"] = max_profit

        #Define items, create multiple columns and scale them individually
        items = pd.DataFrame(lines[5:])
        scaleditems = build_items(items)
        scaleditems = scale_items(scaleditems)
        
        #Write the content of the items DataFrame to the overview DataFrame
        for index in scaleditems.index:
            profit = scaleditems.at[index,"profit"]
            weight = scaleditems.at[index,"weight"]
            write_profit_weight_to_csv(instances_overview, filepath, index, profit, weight)
        print(str(filepath)+" added") #Output to keep track of the progress
    else: pass

#Print information about the DataFrame and save it as instances.csv. So far, all 20,300 instances are included
print(instances_overview.info())
print(instances_overview.describe())       
instances_overview.to_csv("knapsacksolver/scores/instances.csv", index=True, sep=",", header=True, index_label=None, na_rep="")
```

```python
#Preparation of Instances for AAS by removing unsolved instances from the data
#Based on the scores, instances are dismissed
scores = pd.DataFrame(pd.read_csv("knapsacksolver/scores/par10_scores.csv", header=0, index_col=0))
scores = scores.rename(columns={"dynamic_programming_bellman_array":"dynamic_programming", "branch_and_bound_sort": "branch_and_bound"})

#Load instances
data = pd.DataFrame(pd.read_csv("knapsacksolver/scores/instances.csv", header=0, index_col=0))

#Sort data so that indices match
scores = scores.sort_index()
data = data.sort_index()

#Define which rows to remove
unsolved_instances = scores[scores.min(axis=1) == 50000]

#Drop unsolved instances from the scores and the data
scores.drop(unsolved_instances.index, inplace=True)
data.drop(unsolved_instances.index, inplace=True)

#Save each in the folder "AAS" so that the original scores and instances are not overwritten
scores.to_csv("AAS/scores.csv", index=True, sep=",", header=True, index_label=None, na_rep="")
data.to_csv("AAS/instances.csv", index=True, sep=",", header=True, index_label=None, na_rep="")
```

```python
#IMPORTANT NOTE: This approach scales the capacity with the weights and the max_profit with the profits

#Load already scaled data to use the index as solved instances
aas_data = pd.DataFrame(pd.read_csv("AAS/instances.csv", header=0, index_col=0))

#Functions are already used and unchanged
def read_profit(line):
    line = str(line)
    line = line[(line.find(",")+1):]
    line = line[:line.find(",")]
    return int(line)

def read_weight(line):
    line = str(line)
    line = line[(line.find(",")+1):]
    line = line[(line.find(",")+1):]
    line = line[:line.find(",")]
    return int(line)

def initialize_dataframe_columns():
    df = pd.DataFrame(columns=["capacity", "max_profit"])
    for i in range(1,1001):
        newcol = pd.DataFrame(columns=["profit_item_"+str(i), "weight_item_"+str(i)])
        df = pd.concat([df, newcol])
    return df

def build_items(items):
    for index in items.index:
        profit = read_profit(items[0][index])
        weight = read_weight(items[0][index])
        items.at[index, "profit"] = profit
        items.at[index, "weight"] = weight
    return pd.concat([items["profit"], items["weight"]], axis=1)

def scale_items(items):
    min_max_scaler = MinMaxScaler()
    scaled = min_max_scaler.fit_transform(items)
    return pd.DataFrame(scaled, columns=["profit", "weight"])
    
def write_profit_weight_to_csv(df, filepath, index, profit, weight):
    df.at[filepath, "profit_item_"+str(index + 1)] = profit
    df.at[filepath, "weight_item_"+str(index + 1)] = weight

instances_overview = initialize_dataframe_columns()
counter = 1 #Added a counter to better keep track of the progress
for filepath in aas_data.index: #Iterate over every solved isntance
    with open("knapsacksolver/dataset/"+str(filepath)) as file: #Load the original data
        lines = file.readlines()
    capacity = lines[2]
    capacity = int(capacity[2:])
    max_profit = lines[3]
    max_profit = int(max_profit[2:])

    items = pd.DataFrame(lines[5:])
    items = build_items(items) 
    #Items are a DataFrame of the two columns "Profit" and "Weight"
    
    #Add the capacity and max_profit to the items DataFrame
    items.at["c_z", "weight"] = capacity
    items.at["c_z", "profit"] = max_profit
    
    #Scale the columns
    scaleditems = scale_items(items)

    
    for index in scaleditems.index:
        if index == "c_z": #If statement did not work in the intended way to seperate the max_profit and capacity
            instances_overview.at[filepath, "capacity"] = scaleditems.at["c_z","weight"]
            instances_overview.at[filepath, "max_profit"] = scaleditems.at["c_z","profit"]
        else: 
            #Write the properties into the DataFrame
            profit = scaleditems.at[index,"profit"]
            weight = scaleditems.at[index,"weight"]
            write_profit_weight_to_csv(instances_overview, filepath, index, profit, weight)
    
    #Output to keep track
    print(str(filepath)+" added "+str(counter)+"/18633")
    counter += 1 #Increase the counter

print(instances_overview.info())

#Save instances in a differnt file for the scaling approach
instances_overview.to_csv("AAS/instances_to_c_z.csv", index=True, sep=",", header=True, index_label=None, na_rep="")
```
