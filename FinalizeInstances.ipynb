```python
#Function to read max_profit from the command line output
def read_max_profit(string):
    string = string[string.find("Value"):] #Cut everything before "Value" in the output String 
    string= string[:string.find("\\n")] #Cut everything after new line ("\\n will be cut too")
    string = int(string[6:]) #Remove the "Value:" from the String and convert it to an integer
    return string

#Function to read time from the command line output
def read_time(string):
    string = string[string.find("Time"):] #Find the Time output
    string= string[:string.find("\\n")]
    string = float(string[10:]) #As there are many empty spaces after time, they are cut
    return string

#Solve instance to read max_profit and time consumed
def solve_profit_time(filepath):
    #Instances are solved with the algorithm "combo", as it guarantees the optimal solution in short time
    solution = subprocess.run(["./bazel-bin/knapsacksolver/main -v 1 --algorithm combo --input " + filepath + " --format pisinger"], shell=True, capture_output=True)
    solution = str(solution.stdout)
    new_max_profit = read_max_profit(solution)
    new_time = read_time(solution)
    return(new_max_profit, new_time)
```

```python
#Function to overwrite profit and time
def set_new_values(filepath, new_max_profit, new_time):
    #Read instance as lines
    with open(filepath) as file:
        lines = file.readlines()
    #Overwrite the max_profit
    lines[3] = "z "+str(new_max_profit)+"\n"
    #Overwrite the time (This is not necessary for the project but the function to solve the time is important later)
    lines[4] = "time "+str(new_time)+"\n" 
    
    #Write all lines (including the overwritten ones) to the file
    with open(filepath, "w") as file: 
        for line in lines:
            file.write(line)
```

```python
#Solving and overwriting all instances
for filename in os.listdir("dataset/"):
    if "checkpoint" in filename: pass #Avoid Jupyter checkpoints
    else:
        filepath = "dataset/"+filename
        new_max_profit, new_time = solve_profit_time(filepath) #Solve the instances and read the new values
        set_new_values(filepath, new_max_profit, new_time) #Write the new values to the instances
        print(str(filepath)+" finalized.") #Output to review the progress
```
